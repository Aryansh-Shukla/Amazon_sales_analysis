```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

The above graph shows that T-shirts is one product in great demand and bought extensively by the user

```python
B2B_Check=df['B2B'].value_counts()

plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

The pie chart resembles that 99.2% sales are B2C while 0.8% are B2B

```python
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter plot of Category vs Size')
plt.show()
```

Scatter Plot for availability of product along with sizes

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('Ship-State')
plt.ylabel('Count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

State-Wise shipping and Transit data

```python
top_10_states=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('Ship-State')
plt.ylabel('Count')
plt.title('Top 10 states with highest sales')
plt.xticks(rotation=45)
plt.show()
```

Maharashtra emerges as the state with the maximum no. of sales followed by Karnataka and Tamil Nadu.
